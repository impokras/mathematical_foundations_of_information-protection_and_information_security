```julia
function calc_func_x(x, p, a, b)
    if x < p ÷ 2
        return mod(a*x, p)
    else
        return mod(b*x, p)
    end
end

function calc_func_uv(c, p, u, v)
    if c < p ÷ 2
        u=u+1
    else
        v=v+1  
    end 
    return u, v
end


function pollards_rho_log(p, a, b, u1, v1, u2, v2)

    
    c = (a^u1 % p) * (b^v1 % p) % p
    d = c

    println("с: $c - $u1 + $v1 x")
    println("d: $d - $u2 + $v2 x")

    u1, v1 = calc_func_uv(c, p, u1, v1)
    c = calc_func_x(c, p, a, b)

    u2, v2 = calc_func_uv(d, p, u2, v2)
    d = calc_func_x(d, p, a, b)

    u2, v2 = calc_func_uv(d, p, u2, v2)
    d = calc_func_x(d, p, a, b)

    println("с: $c - $u1 + $v1 x")
    println("d: $d - $u2 + $v2 x")

    while c != d
        
        u1, v1 = calc_func_uv(c, p, u1, v1)
        c = calc_func_x(c, p, a, b)

        u2, v2 = calc_func_uv(d, p, u2, v2)
        d = calc_func_x(d, p, a, b)
        
        u2, v2 = calc_func_uv(d, p, u2, v2)
        d = calc_func_x(d, p, a, b)
        println("с: $c - $u1 + $v1 x")
        println("d: $d - $u2 + $v2 x")
    end

    x = 1

    while mod((v1-v2)*x, p ÷ 2) != mod((u2-u1), p ÷ 2)
        x+=1
    end

    println("x = $x") 
end

p = 107
a = 10
b = 64

pollards_rho_log(p, a, b, 2, 2, 2, 2)
```
